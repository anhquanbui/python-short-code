
# Python Dictionary Methods â€“ Hands-on Examples

This notebook demonstrates **all common Python dictionary methods** with clear,
runnable examples.

Covered methods:
- clear()
- copy()
- fromkeys()
- get()
- items()
- keys()
- pop()
- popitem()
- setdefault()
- update()
- values()


## 1) Setup: sample dictionary

```python

car = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

print(car)
```

## 2) clear() â€“ remove all items

```python

d = car.copy()
d.clear()
print(d)
```

## 3) copy() â€“ shallow copy

```python

d1 = car.copy()
d1["year"] = 2024

print("original:", car)
print("copy:", d1)
```

## 4) fromkeys() â€“ create dict from keys

```python

keys = ["a", "b", "c"]

d = dict.fromkeys(keys, 0)
print(d)
```

## 5) get() â€“ safe access

```python

print(car.get("brand"))
print(car.get("engine"))          # None
print(car.get("engine", "N/A"))   # default
```

## 6) items() â€“ key/value pairs

```python

for k, v in car.items():
    print(k, "=>", v)
```

## 7) keys() â€“ dictionary keys

```python

print(car.keys())
print(list(car.keys()))
```

## 8) values() â€“ dictionary values

```python

print(car.values())
print(list(car.values()))
```

## 9) pop() â€“ remove key and return value

```python

d = car.copy()
year = d.pop("year")

print("popped:", year)
print("after pop:", d)
```

## 10) popitem() â€“ remove last inserted item

```python

d = car.copy()
item = d.popitem()

print("removed:", item)
print("after popitem:", d)
```

## 11) setdefault() â€“ get or insert

```python

d = car.copy()

engine = d.setdefault("engine", "V8")
year = d.setdefault("year", 2000)

print("engine:", engine)
print("year:", year)
print("dict:", d)
```

## 12) update() â€“ merge key-value pairs

```python

d = car.copy()

d.update({
    "year": 2025,
    "color": "red"
})

print(d)
```
