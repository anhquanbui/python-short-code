```python
print("=== CASE 1: SHARED REFERENCE (BUG) ===")

video = {
    "id": "abc123",
    "statistics": {"views": 100},
    "tags": ["music"]
}

doc = video              # ❌ không copy

doc["statistics"]["views"] += 50
doc["tags"].append("live")

print("video:", video)
print("doc:  ", doc)
```

```python
print("\n=== CASE 2: SHALLOW COPY (BUG NGẦM) ===")

video = {
    "id": "abc123",
    "statistics": {"views": 100},
    "tags": ["music"]
}

doc = video.copy()        # ❌ shallow copy

doc["statistics"]["views"] += 50
doc["tags"].append("live")

print("video:", video)
print("doc:  ", doc)
```

```python
print("\n=== CASE 3: DEEP COPY (AN TOÀN) ===")

import copy

video = {
    "id": "abc123",
    "statistics": {"views": 100},
    "tags": ["music"]
}

doc = copy.deepcopy(video)

doc["statistics"]["views"] += 50
doc["tags"].append("live")

print("video:", video)
print("doc:  ", doc)
```

```python
print("\n=== CASE 4: SELECTIVE COPY (CHUẨN DÙNG THỰC TẾ) ===")

video = {
    "id": "abc123",
    "statistics": {"views": 100},
    "tags": ["music"],
    "ml_flags": {"low_quality": False}
}

doc = video.copy()                         # copy vỏ
doc["statistics"] = video["statistics"].copy()
doc["tags"] = list(video["tags"])
doc["ml_flags"] = video["ml_flags"].copy()

doc["statistics"]["views"] += 50
doc["tags"].append("live")
doc["ml_flags"]["low_quality"] = True

print("video:", video)
print("doc:  ", doc)
```
